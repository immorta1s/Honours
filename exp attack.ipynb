# CHi-Square test

```python
from sklearn.feature_selection import chi2
from sklearn.feature_selection import SelectKBest
import matplotlib.pyplot as plt

predictors = refclasscol
X_chi = SelectKBest(chi2, k=15).fit(x1, y_use)
scores= X_chi.scores_
#print(scores)
rank= X_chi.get_support(True)
for i in rank:
    print(refclasscol[i])
new_X_chi = X_chi.transform(x1)
# Plot the scores.  See how "Pclass", "Sex", "Title", and "Fare" are the best?
plt.figure(figsize=(15,10))
plt.bar(range(len(predictors)), scores)
plt.xticks(range(len(predictors)), predictors, rotation='vertical')
plt.show()
```

# RFE test

```python
from sklearn.feature_selection import RFE
from sklearn.ensemble import RandomForestClassifier
import itertools

rfc = RandomForestClassifier()

# create the RFE model and select 15 attributes
rfe = RFE(rfc, n_features_to_select=15,step=1)
rfe = rfe.fit(x1, y_use)

# summarize the selection of the attributes
feature_map = [(i, v) for i, v in itertools.zip_longest(rfe.get_support(), refclasscol)]
selected_features = [v for i, v in feature_map if i==True]
```

```python
new_X_rfe = rfe.transform(x1)
```

```python
import matplotlib.pyplot as plt
scores = rfe.ranking_
predictors = refclasscol
plt.figure(figsize=(15,10))
plt.bar(range(len(predictors)), scores)
#plt.xlabel('Feature')
plt.ylabel('Ranking')
plt.xticks(range(len(predictors)), predictors, rotation='vertical')
plt.show()
```
